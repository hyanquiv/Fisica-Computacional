```python
import matplotlib.pyplot as plt
import numpy as np

# Constantes
m = 1.0  # masa en kg
g = 9.81  # aceleración debida a la gravedad en m/s^2
h_inicial = 10.0  # altura inicial en metros
v_inicial = 0.0  # velocidad inicial en m/s

# Tiempo y parámetros
t = np.linspace(0, 2, 100)  # tiempo de 0 a 2 segundos con 100 puntos
h = h_inicial - 0.5 * g * t**2  # altura en función del tiempo
v = -g * t  # velocidad en función del tiempo

# Energías
E_potencial = m * g * h
E_cinetica = 0.5 * m * v**2
E_mecanica = E_potencial + E_cinetica

# Gráficas
plt.figure(figsize=(10, 6))
plt.plot(t, E_potencial, label='Energía Potencial (Ep)')
plt.plot(t, E_cinetica, label='Energía Cinética (Ek)')
plt.plot(t, E_mecanica, label='Energía Mecánica (Em)', linestyle='--')

plt.xlabel('Tiempo (s)')
plt.ylabel('Energía (J)')
plt.title('Conservación de la Energía Mecánica')
plt.legend()
plt.grid(True)
plt.show()
```
