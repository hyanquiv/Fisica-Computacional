## Ejercicio n° 1

```python
import numpy as np
from ipywidgets import interactive
import ipywidgets as widgets

# Parámetros comunes
L = 10           # Longitud de la barra
Nx = 100         # Número de puntos en el espacio
dx = L / (Nx - 1)
x = np.linspace(0, L, Nx)

# Función para resolver la ecuación de calor
def ecuacion_calor(Tmax, dt, alpha):
    Nt = int(Tmax / dt)
    u = np.zeros((Nt, Nx))  # Matriz para almacenar resultados
    r = alpha * dt / dx**2

    # Condición inicial (ejemplo: distribución de temperatura en forma de seno)
    u[0, :] = np.sin(np.pi * x / L)

    # Iteración en el tiempo
    for j in range(0, Nt - 1):
        for i in range(1, Nx - 1):
            u[j + 1, i] = (1 - 2 * r) * u[j, i] + r * (u[j, i + 1] + u[j, i - 1])

    return u[-1, :]  # Retorna solo el resultado final (último tiempo)

# Función para mostrar el resultado de la ecuación de calor
def mostrar_calor(Tmax=1.0, dt=0.01, alpha=0.01):
    u = ecuacion_calor(Tmax, dt, alpha)
    print("Resultado actual de la ecuación de calor (último estado temporal):")
    print(u)

# Interfaz interactiva para la ecuación de calor
interactive_calor = interactive(mostrar_calor, Tmax=widgets.FloatSlider(min=0.1, max=5.0, step=0.1, value=1.0),
                                dt=widgets.FloatSlider(min=0.001, max=0.1, step=0.001, value=0.01),
                                alpha=widgets.FloatSlider(min=0.001, max=0.1, step=0.001, value=0.01))

# Visualización
print("Ecuación de Calor:")
display(interactive_calor)
```

## Ejercicio n° 2


```python
import numpy as np
from ipywidgets import interactive
import ipywidgets as widgets

# Parámetros comunes
L = 10           # Longitud de la cuerda
Nx = 100         # Número de puntos en el espacio
dx = L / (Nx - 1)
x = np.linspace(0, L, Nx)

# Función para resolver la ecuación de onda
def ecuacion_onda(Tmax, dt, c):
    Nt = int(Tmax / dt)
    u = np.zeros((Nt, Nx))  # Matriz para almacenar resultados
    r = c * dt / dx

    # Condición inicial (ejemplo: perturbación inicial)
    u[0, :] = np.sin(np.pi * x / L)
    u[1, :] = u[0, :]  # Aproximación para el segundo paso en t

    # Iteración en el tiempo
    for j in range(1, Nt - 1):
        for i in range(1, Nx - 1):
            u[j + 1, i] = 2 * (1 - r**2) * u[j, i] + r**2 * (u[j, i + 1] + u[j, i - 1]) - u[j - 1, i]

    return u[-1, :]  # Retorna solo el resultado final (último tiempo)

# Función para mostrar el resultado de la ecuación de onda
def mostrar_onda(Tmax=1.0, dt=0.01, c=1.0):
    u = ecuacion_onda(Tmax, dt, c)
    print("Resultado actual de la ecuación de onda (último estado temporal):")
    print(u)

# Interfaz interactiva para la ecuación de onda
interactive_onda = interactive(mostrar_onda, Tmax=widgets.FloatSlider(min=0.1, max=5.0, step=0.1, value=1.0),
                               dt=widgets.FloatSlider(min=0.001, max=0.1, step=0.001, value=0.01),
                               c=widgets.FloatSlider(min=0.1, max=10.0, step=0.1, value=1.0))

# Visualización
print("Ecuación de Onda:")
display(interactive_onda)
```
